````python
# pls stfu tensorflow
import warnings, os
warnings.filterwarnings("ignore")
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

from tensorflow import compat
compat.v1.logging.set_verbosity(compat.v1.logging.FATAL)

# import
from addressnet.predict import predict_one
import pandas as pd
````

````python
df = pd.DataFrame(columns=['raw', 'parse']).fillna('')
df['raw'] = ['65 Parkes Road', 
             'university of melbourne parkville victoria',
             '13 fake street nsw',
             '22 Marloo Street MARDEN South Australia 5070',
             '16 Boorie Road Brigooda Queensland',
             '6 Shaw Drive,3645,craigie,victoria,au',
             '98 Shirley Street PIMPAMA QLD 4209',
             'Flat 2 14 Smith St',
             '2/14 will St',
             'PO Box 105 North Sydney NSW 2060 AUSTRALIA'] # po boxes dont really work
````

````python
%%capture
# this magic cell gets rid of any bs tensorflow warning lmao
df['parse'] = df['raw'].apply(predict_one)
````
